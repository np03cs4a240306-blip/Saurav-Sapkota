```python
import numpy as np

def gradient_descent_step(X, Y, W, alpha):

    if W.ndim == 1:
        W = W.reshape(-1, 1)
    if Y.ndim == 1:
        Y = Y.reshape(-1, 1)

    m = len(Y)
    H = X @ W
    Error = H - Y
    Gradient = (1/m) * (X.T @ Error)
    W_new = W - (alpha * Gradient)

    return W_new

X_test = np.array([[1, 2], [3, 4], [5, 6]])
Y_test = np.array([3, 7, 11]).reshape(-1, 1) # Reshape to (3, 1)
W_initial = np.array([0.5, 0.5]).reshape(-1, 1)  # Initial guess for W, shape (2, 1)
alpha_rate = 0.01

W_new = gradient_descent_step(X_test, Y_test, W_initial, alpha_rate)

print("--- Gradient Descent Step Verification ---")
print(f"Initial W:\n{W_initial.T}")
print(f"Learning Rate (alpha): {alpha_rate}")
print(f"Updated W:\n{W_new.T}")
```

```python
import numpy as np
def gradient_descent(X, Y, W, alpha, iterations):
    if W.ndim == 1:
        W = W.reshape(-1, 1)
    if Y.ndim == 1:
        Y = Y.reshape(-1, 1)

    cost_history = [0] * iterations
    m = len(Y)
    W_update = W

    for iteration in range(iterations):
        Y_pred = X @ W_update
        loss = Y_pred - Y
        dw = (1/m) * (X.T @ loss)
        W_update = W_update - (alpha * dw)
        cost = cost_function(X, Y, W_update)
        cost_history[iteration] = cost

    return W_update, cost_history
```

```python
import numpy as np

# --- Cost Function (from To-Do-4) ---
def cost_function(X, Y, W):
    if W.ndim == 1:
        W = W.reshape(-1, 1)
    if Y.ndim == 1:
        Y = Y.reshape(-1, 1)
    m = len(Y)
    H = X @ W
    Error = H - Y
    Sum_Squared_Error = Error.T @ Error
    cost = Sum_Squared_Error[0, 0] / (2 * m)
    return cost

# --- Gradient Descent Function (from To-Do-6) ---
def gradient_descent(X, Y, W, alpha, iterations):
    # Ensure W and Y are column vectors for robust matrix operations
    if W.ndim == 1:
        W = W.reshape(-1, 1)
    if Y.ndim == 1:
        Y = Y.reshape(-1, 1)

    cost_history = [0] * iterations
    m = len(Y)
    W_update = W

    for iteration in range(iterations):
        # Step 1: Hypothesis Values (Prediction: H = X @ W)
        Y_pred = X @ W_update

        # Step 2: Difference between Hypothesis and Actual Y (Error)
        loss = Y_pred - Y

        # Step 3: Gradient Calculation (Gradient = (1/m) * X.T @ loss)
        dw = (1/m) * (X.T @ loss)

        # Step 4: Updating Values of W using Gradient
        W_update = W_update - (alpha * dw)

        # Step 5: New Cost Value
        cost = cost_function(X, Y, W_update)
        cost_history[iteration] = cost

    return W_update, cost_history

# --- Test Execution ---

# Generate random test data
np.random.seed(0) # For reproducibility
X = np.random.rand(100, 3) # 100 samples, 3 features
Y = np.random.rand(100)
W = np.random.rand(3) # Initial guess for parameters
# Set hyperparameters
alpha = 0.01
iterations = 1000

# Test the gradient_descent function
final_params, cost_history = gradient_descent(X, Y, W, alpha, iterations)

# Print the results
print("--- Gradient Descent Test Results ---")
print(f"Initial W (Guess): {W}")
print(f"Learning Rate (alpha): {alpha}")
print(f"Iterations: {iterations}")
print("-" * 35)
print(f"Final Parameters (Optimized W):\n{final_params.T}")
print(f"Initial Cost: {cost_history[0]:.6f}")
print(f"Final Cost: {cost_history[-1]:.6f}")
```

```python
import numpy as np

def rmse(Y, Y_pred):
    if Y.shape != Y_pred.shape:
        if Y.ndim > Y_pred.ndim:
             Y_pred = Y_pred.reshape(-1, 1)
        elif Y_pred.ndim > Y.ndim:
             Y = Y.reshape(-1, 1)

    m = len(Y)

    squared_error = (Y_pred - Y)**2

    mse = np.sum(squared_error) / m

    rmse = np.sqrt(mse)

    return rmse

# Test data (actual values)
Y_actual = np.array([10, 20, 30])

# Predicted values
Y_predicted = np.array([10.5, 19.8, 31.0])

calculated_rmse = rmse(Y_actual, Y_predicted)

print("--- RMSE Calculation Test ---")
print(f"Actual Y:   {Y_actual}")
print(f"Predicted Y: {Y_predicted}")
print(f"Calculated RMSE: {calculated_rmse:.4f}")
```

```python
import numpy as np

# Model Evaluation - R2
def r2(Y, Y_pred):

    if Y.shape != Y_pred.shape:
        if Y.ndim > Y_pred.ndim:
             Y_pred = Y_pred.reshape(-1, 1)
        elif Y_pred.ndim > Y.ndim:
             Y = Y.reshape(-1, 1)

    mean_y = np.mean(Y)

    ss_tot = np.sum((Y - mean_y)**2)
    ss_res = np.sum((Y - Y_pred)**2)

    # R-Squared Formula
    r2 = 1 - (ss_res / ss_tot)

    return r2
Y_actual = np.array([10, 20, 30, 40])
Y_predicted_good = np.array([11, 21, 29, 39])
Y_predicted_poor = np.array([25, 25, 25, 25])

r2_good = r2(Y_actual, Y_predicted_good)
r2_poor = r2(Y_actual, Y_predicted_poor)

print("--- R-Squared Calculation Test ---")
print(f"R2 for Good Fit (closer to 1.0): {r2_good:.4f}")
print(f"R2 for Poor Fit (closer to 0.0): {r2_poor:.4f}")
```

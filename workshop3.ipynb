```python
import pandas as pd

# Sample DataFrame
mydataframe = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 20],
    'Score': [88, 92, 85]
})

# Sorting by values
sorted_by_age = mydataframe.sort_values(by='Age', ascending=True, inplace=False)
print("Sorted by Age:\n", sorted_by_age)

# Sorting by index
sorted_by_index = mydataframe.sort_index(ascending=True, inplace=False)
print("\nSorted by Index:\n", sorted_by_index)
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 19, 22, 25],
    'Score': [88, 92, 85, 95]
}

df = pd.DataFrame(data)
print(df.head())

# Example 1: Sort by 'Age' using sort_values()
sorted_by_age = df.sort_values(by='Age')
print("\nSorted by Age:\n", sorted_by_age.head())

# Example 2: Sort by index using sort_index()
sorted_by_index = df.sort_index()
print("\nSorted by Index:\n", sorted_by_index.head())
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 19, 22, 25],
    'Score': [88, 92, 85, 95]
}

df = pd.DataFrame(data)

# 1. Using iloc[]: Accessing rows and columns by index
subset_iloc = df.iloc[1:3, 0:2]
print("Using iloc[]:\n", subset_iloc)

# 2. Using loc[]: Accessing rows by condition and specific columns
subset_loc = df.loc[df['Age'] > 20, ['Name', 'Score']]
print("\nUsing loc[] (Age > 20):\n", subset_loc)

# 3. Using []: Selecting specific columns
subset_brackets = df[['Name', 'Age']]
print("\nSelecting specific columns:\n", subset_brackets)
```

```python
import pandas as pd

# df: DataFrame from slide 09
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 19, 22, 25],
    'Score': [88, 92, 85, 95]
}

df = pd.DataFrame(data)

# Filter rows with a single condition (Age > 20)
filtered_single = df[df['Age'] > 20]
print("Single Condition (Age > 20):\n", filtered_single)

# Filter rows with multiple conditions (Age > 20 and Score > 85)
filtered_multiple = df[(df['Age'] > 20) & (df['Score'] > 85)]
print("\nMultiple Conditions (Age > 20 AND Score > 85):\n", filtered_multiple)
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'HR'],
    'Score': [88, 92, 85, 95]
}

df = pd.DataFrame(data)

# Filter rows where 'Department' is 'HR'
filtered_single = df[df['Department'] == 'HR']
print("Department = HR:\n", filtered_single)

# Filter rows where 'Department' is either 'HR' or 'IT'
filtered_multiple = df[df['Department'].isin(['HR', 'IT'])]
print("\nDepartment = HR or IT:\n", filtered_multiple)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

# Aggregation: Calculate mean for each group
grouped = df.groupby('Category')['Value'].mean()
print(grouped)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

# Transformation: Normalize values within each group
df['Normalized'] = df.groupby('Category')['Value'].transform(lambda x: x / x.sum())

print(df)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

# Filtration: Keep groups where sum of values > 60
filtered = df.groupby('Category').filter(lambda x: x['Value'].sum() > 60)

print(filtered)
```

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]

# Plot this week
plt.plot(days, steps_walked, marker='o', color='green', linestyle='-', label='This week')

# Plot last week
plt.plot(days, steps_last_week, marker='v', color='magenta', linestyle='--', label='Last week')

# Titles and labels
plt.title("Step Count | This Week and Last Week")
plt.xlabel("Days of the Week")
plt.ylabel("Steps Walked")

# Grid and legend
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```
